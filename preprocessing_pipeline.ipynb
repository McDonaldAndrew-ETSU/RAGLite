```python
def preprocess_corrupted_json_stream(input_file, output_file):
    buffer = ""  # Holds potential JSON objects
    brace_stack = []  # Stack to track nested object depth
    valid_objects = []  # Stores cleaned JSON objects

    with open(input_file, "r", encoding="utf-8") as f:
        while chunk := f.read(4096):  # Read in small chunks (4KB)
            for char in chunk:
                buffer += char  # Append character to buffer

                # Track nesting level
                if char == "{":
                    brace_stack.append("{")
                elif char == "}":
                    if brace_stack:
                        brace_stack.pop()

                # When brace_stack is empty, we have a complete object
                if not brace_stack and buffer.strip():
                    cleaned_buffer = re.sub(
                        r"\bNaN\b|\bNan\b|\bNAN\b", "null", buffer
                    )  # Replace NaNs

                    try:
                        json_obj = json.loads(cleaned_buffer)  # Attempt to parse JSON
                        valid_objects.append(json_obj)  # Store valid object
                    except json.JSONDecodeError as e:
                        print(f"Skipping corrupted JSON object. Error: {e}")

                    buffer = ""  # Reset buffer for next JSON object

    # Write cleaned JSON objects to output file
    with open(output_file, "w", encoding="utf-8") as f_out:
        json.dump(valid_objects, f_out, indent=4)
        print(f"Successfully cleaned and saved JSON to {output_file}")
```

```python
source_dir = "./data/Students"
cleaned_dir = "./data/cleaned_Students"

os.makedirs(cleaned_dir, exist_ok=True)

for file_name in os.listdir(source_dir):
    input_path = os.path.join(source_dir, file_name)
    output_path = os.path.join(cleaned_dir, f"cleaned_{file_name}")

    preprocess_corrupted_json_stream(input_path, output_path)

print("All files pre-processed")
```
